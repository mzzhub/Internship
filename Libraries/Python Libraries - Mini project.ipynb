# Visualization

## Gender

```python
# Count plot for "F" and "M" in column "Gender"

sns.set(rc = {"figure.figsize" : (5,5)})
for_count = sns.countplot(data = df, x = "Gender", hue = "Gender")

for i in for_count.containers:
    for_count.bar_label(i)
```

```python
# Bar plot on 'total amount spend' by females and males

amount_on_gender = df.groupby(["Gender"], as_index = False)["Amount"].sum()

sns.set(rc = {"figure.figsize" : (6,5)})
for_count = sns.barplot(data = amount_on_gender, x = "Gender", y = "Amount", hue = "Gender")

for i in for_count.containers:
    for_count.bar_label(i)
```

_From above graphs, we can see that most of the buyers are females and even the purchasing power of females are greater than men._

## Age

```python
# Count plot on different entries in column "Age Group"

queue = ["0-17", "18-25", "26-35", "36-45", "46-50", "51-55", "55+"]

sns.set(rc = {"figure.figsize" : (5,5)})
for_count = sns.countplot(data = df, x = "Age Group", hue = "Age Group", order = queue)

for i in for_count.containers:
    for_count.bar_label(i)
```

```python
# Count plot on different entries in column "Age Group". With 'female and male' seperated.

queue = ["0-17", "18-25", "26-35", "36-45", "46-50", "51-55", "55+"]

sns.set(rc = {"figure.figsize" : (15,5)})
for_count = sns.countplot(data = df, x = "Age Group", hue = "Gender", palette="dark:red", order = queue)

for i in for_count.containers:
    for_count.bar_label(i)
```

```python
# Bar plot on 'total amount spend' on "Age group"
amount_on_ageGroup = df.groupby(["Age Group"], as_index = False)["Amount"].sum()

sns.set(rc = {"figure.figsize" : (15,5)})
for_count = sns.barplot(data = amount_on_ageGroup, x = "Age Group", y = "Amount")

for i in for_count.containers:
    for_count.bar_label(i)
```

```python
amount_on_ageGroup
```

```python
# Bar plot on 'total amount spend' on "Age group" for different "Gender"

amount_on_ageGroup = df.groupby(["Age Group", "Gender"], as_index = False)["Amount"].sum()

sns.set(rc = {"figure.figsize" : (15,5)})
for_count = sns.barplot(data = amount_on_ageGroup, x = "Age Group", y = "Amount", hue = "Gender")

for i in for_count.containers:
    for_count.bar_label(i, rotation = 90, padding = 5)
```

_From above graphs, we can see that most of the buyers are females of age group between 26-35 years._

## State

```python
# Bar plot on 'total number of orders' with respect to different "States"

orders_on_state = df.groupby(["State"], as_index = False)["Orders"].sum()

sns.set(rc = {"figure.figsize" : (15,5)})
for_count = sns.barplot(data = orders_on_state, x = "State", y = "Orders")
plt.xticks(rotation = 90)

for i in for_count.containers:
    for_count.bar_label(i)
```

```python
# Same as above, but in descending order.

orders_on_state = df.groupby(["State"], as_index = False)["Orders"].sum().sort_values(by = "Orders", ascending=False)

sns.set(rc = {"figure.figsize" : (15,5)})
for_count = sns.barplot(data = orders_on_state, x = "State", y = "Orders", hue = "State")
plt.xticks(rotation = 90)

for i in for_count.containers:
    for_count.bar_label(i)
```

```python
# Bar plot on 'total amount spend' with respect to different "State"

amount_on_state = df.groupby(["State"], as_index = False)["Amount"].sum()

sns.set(rc = {"figure.figsize" : (15,5)})
for_count = sns.barplot(data = amount_on_state, x = "State", y = "Amount")
plt.xticks(rotation = 90)

for i in for_count.containers:
    for_count.bar_label(i, rotation = 90, padding = 5)
```

```python
# Same as above, but in descending order.

amount_on_state = df.groupby(["State"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

sns.set(rc = {"figure.figsize" : (15,5)})
for_count = sns.barplot(data = amount_on_state, x = "State", y = "Amount", hue = "State")
plt.xticks(rotation = 90)

for i in for_count.containers:
    for_count.bar_label(i, rotation = 90, padding = 5)
```

* _From above graphs we can see that most of the orders & total amount are from Uttar Pradesh, Maharashtra and Karnataka._
* _Kerala is in 8 <sup>th</sup> position in ' total number of orders ' , but in 11 <sup>th</sup> position in ' total amount spend '._

## Marital status

```python
# Count plot on "Matrial_Status"

sns.set(rc = {"figure.figsize" : (5,5)})

for_count = sns.countplot(data = df, x = "Marital_Status")

for_count.set_xticklabels(["Unarried", "Married"])

for i in for_count.containers:
    for_count.bar_label(i)
```

```python
df.groupby(["Marital_Status"], as_index = False)["Marital_Status"].count()
```

```python
# Bar plot on 'total amount spend' with respect to "Marital_Status"

amount_on_marital = df.groupby(["Marital_Status"], as_index = False)["Amount"].sum()

for_count = sns.barplot(data = amount_on_marital, x = "Marital_Status", y = "Amount", hue = "Marital_Status")

for_count.set_xticklabels(["Unarried", "Married"])

for i in for_count.containers:
    for_count.bar_label(i)
```

```python
# Bar plot on 'total amount spend' with respect to "Marital_Status", but seperated by "0" and "1" in "Marital_Status".

amount_on_marital = df.groupby(["Marital_Status", "Gender"], as_index = False)["Amount"].sum()

sns.set(rc = {"figure.figsize" : (10,5)})

for_count = sns.barplot(data = amount_on_marital, x = "Marital_Status", y = "Amount", hue = "Gender", color = "red")

for_count.set_xticklabels(["Unmarried", "Married"])

for i in for_count.containers:
    for_count.bar_label(i)
```

_From above graphs we can see that most of the buyers are married (women)_

## Occupation

```python
# Count plot on "Occupation"

sns.set(rc = {"figure.figsize" : (15,5)})

for_count = sns.countplot(data = df, x = "Occupation")

plt.xticks(rotation = 90)

for i in for_count.containers:
    for_count.bar_label(i)
```

```python
# Same as above, but in descending order.

sns.set(rc = {"figure.figsize" : (15,5)})

queue = df["Occupation"].value_counts().index

for_count = sns.countplot(data = df, x = "Occupation", order = queue, hue = "Occupation")

plt.xticks(rotation = 90)

for i in for_count.containers:
    for_count.bar_label(i)
```

```python
# Bar plot on 'total amount spend' with respect to different "Occupation".

amount_on_occupation = df.groupby(["Occupation"], as_index = False)["Amount"].sum()

sns.set(rc = {"figure.figsize" : (30,10)})

for_count = sns.barplot(data = amount_on_occupation, x = "Occupation", y = "Amount")

plt.xticks(rotation = 90, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel("Occupation", fontsize = 25, fontweight = "bold")
plt.ylabel("Amount", fontsize = 25, fontweight = "bold")

for i in for_count.containers:
    for_count.bar_label(i)
```

```python
# Same as above, but in descending order.

amount_on_occupation = df.groupby(["Occupation"], as_index = False)["Amount"].sum()

sns.set(rc = {"figure.figsize" : (30,10)})

queue = df["Occupation"].value_counts().index

for_count = sns.barplot(data = amount_on_occupation, x = "Occupation", y = "Amount", order = queue, hue = "Occupation")

plt.xticks(rotation = 90, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel("Occupation", fontsize = 25, fontweight = "bold")
plt.ylabel("Amount", fontsize = 25, fontweight = "bold")

for i in for_count.containers:
    for_count.bar_label(i)
```

* _From above graphs, we can see that most of the buyers are working in IT, Healthcare and Aviation sector._
* _There is no change in the order of heirarchy between count and actual power._

## Product category

```python
df["Product_Category"].unique()
```

```python
len(df["Product_Category"].unique())
```

```python
# Count plot on "Product_Category"

for_count = sns.countplot(data = df, x = "Product_Category")

for i in for_count.containers:
    for_count.bar_label(i, fontsize = 15)

plt.xlabel("Product Category", fontsize = 25, fontweight = "bold")
plt.ylabel("Count", fontsize = 25, fontweight = "bold")
plt.xticks(fontsize = 20,rotation = 90)
plt.yticks(fontsize = 20)

plt.show()
```

```python
# Same as above, but in descending order.

queue = df["Product_Category"].value_counts().index

for_count = sns.countplot(data = df, x = "Product_Category", order = queue, hue = "Product_Category")

for i in for_count.containers:
    for_count.bar_label(i, fontsize = 15)

plt.xlabel("Product Category", fontsize = 25, fontweight = "bold")
plt.ylabel("Count", fontsize = 25, fontweight = "bold")
plt.xticks(fontsize = 20,rotation = 90)
plt.yticks(fontsize = 20)
plt.show()
```

```python
# Bar plot on 'total amount spend' with respect to "Product_Category"

sns.set(rc = {"figure.figsize" : (30,10)})

amount_on_productCategory = df.groupby(["Product_Category"], as_index = False)["Amount"].sum()

for_count = sns.barplot(data = amount_on_productCategory, x = "Product_Category", y = "Amount")

for i in for_count.containers:
    for_count.bar_label(i)

plt.xticks(fontsize = 20, rotation = 90)
plt.yticks(fontsize = 20)
plt.xlabel("Product category", fontsize = 25, fontweight = "bold")
plt.ylabel("Amount", fontsize = 25, fontweight = "bold")
plt.show()
```

```python
# Same as above, but in descending order

sns.set(rc = {"figure.figsize" : (30,10)})

amount_on_productCategory = df.groupby(["Product_Category"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

for_count = sns.barplot(data = amount_on_productCategory, x = "Product_Category", y = "Amount", hue = "Product_Category")

for i in for_count.containers:
    for_count.bar_label(i)

plt.xticks(fontsize = 20, rotation = 90)
plt.yticks(fontsize = 20)
plt.xlabel("Product category", fontsize = 25, fontweight = "bold")
plt.ylabel("Amount", fontsize = 25, fontweight = "bold")
plt.show()
```

```python
# Same as above, but only the first 5 bars

sns.set(rc = {"figure.figsize" : (10,8)})

amount_on_productCategory = df.groupby(["Product_Category"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False).head()

for_count = sns.barplot(data = amount_on_productCategory, x = "Product_Category", y = "Amount", hue = "Product_Category")

for i in for_count.containers:
    for_count.bar_label(i)

plt.xticks(rotation = 90)
plt.xlabel("Product category", fontsize = 25, fontweight = "bold")
plt.ylabel("Amount", fontsize = 25, fontweight = "bold")
plt.show()
```

* _From above graphs, we can see that most of the sold products are from Food, Clothing and Electronics category_
* _Eventhough 'Food' is in secont position on countplot, but it is far more higher when compared to the bar plot of amount._
* _also for 'furniture'._
