# `pandas.concat()`

```python
import pandas as pd
```

```python
one = pd.DataFrame(
    {
        'Name':['Alex','Amy','Allen','Alice','Ayoung','sarang'],
        'subject_id':['sub1','sub2','sub4','sub6','sub5','sub6'],
        'Marks_scored':[98,90,87,67,78,87]
    },
    index=[1,2,3,4,5,6])


two = pd.DataFrame(
    {
        'Name':['Billy','Brian','Bran','Bryce','Betty'],
        'subject_id':['sub2','sub4','sub3','sub6','sub5'],
        'Marks_scored':[89,80,79,97,88]
    },
    index=[1,2,3,4,5])


three = pd.DataFrame(
    {
        'Name':['Billy','Brian','Bran','Bryce','Betty'],
        'subject_id':['sub2','sub4','sub3','sub6','sub5'],
        'Marks_scored':[89,80,79,97,88],
        'roll no':[1,23,4,5,5]
    },
    index=[1,2,3,4,5])
```

```python
one
```

```python
two
```

```python
three
```

```python
df = pd.concat([one,two])
```

```python
df
```

```python
# pandas.append() method is removed in the latest update
```

```python
df = pd.concat([one,two], keys = ["One","Two"])
```

```python
df
```

```python
df = pd.concat([one,two], keys = ["One", "Two"], ignore_index = True)
df
```

```python
df = pd.concat([one,two], keys = ["One", "Two"], ignore_index = True, axis = 1) # default axis = 0
df
```

```python
df = pd.concat([one,two], keys = ["ONE", "TWO"], axis = 1) # now the column headers are back
df
```

```python
df = pd.concat([one,three], axis = 1)
df
```

# `pandas.merge()`

```python
two = pd.DataFrame(
    {
        'id':[1,2,3,4,5],
        'Name':['Billy','Brian','Bran','Bryce','Betty'],
        'subject_id':['sub2','sub4','sub3','sub6','sub5'],
        'Marks_scored':[89,80,79,97,88]
    },
    index=[1,2,3,4,5])


three = pd.DataFrame(
    {
        'id':[1,2,3,4,5],
        'Name':['Billy','Brian','Bran','Bryce','Betty'],
        'subject_id':['sub2','sub4','sub3','sub6','sub5'],
        'Marks_scored':[89,80,79,97,88],
        'roll no':[1,23,4,5,5]
    },
    index=[1,2,3,4,5])
```

```python
two
```

```python
three
```

```python
df = pd.merge(two, three, on = "id")
df
```

```python
two = pd.DataFrame(
    {
        'id':[1,2,3,4,5],
        'Name':['Billy','Brian','Bran','Bryce','Betty'],
        'subject_id':['sub2','sub4','sub3','sub6','sub5'],
        'Marks_scored':[89,80,79,97,88]
    },
    index=[1,2,3,4,5])


three = pd.DataFrame(
    {
        'id':[1,2,3,4,5,6],
        'Name':['Billy','Brian','Bran','Bryce','Betty','Peter'],
        'subject_id':['sub2','sub4','sub3','sub6','sub5','sub6'],
        'Marks_scored':[89,80,79,97,88,67],
        'roll no':[1,23,4,5,5,6]
    },
    index=[1,2,3,4,5,6])
```

```python
df = pd.merge(two, three, on = "id")
df
```

```python
df = pd.merge(two, three, on = ["id", "subject_id"])
df
```

```python
two = pd.DataFrame(
    {
        'id':[1,2,3,4,5,7],
        'Name':['Billy','Brian','Bran','Bryce','Betty','Butcher'],
        'subject_id':['sub2','sub4','sub3','sub6','sub5','sub7'],
        'Marks_scored':[89,80,79,97,88,78]
    },
    index=[1,2,3,4,5,6])


three = pd.DataFrame(
    {
        'id':[1,2,3,4,5,6],
        'Name':['Billy','Brian','Bran','Bryce','Betty','Peter'],
        'subject_id':['sub2','sub4','sub3','sub6','sub5','sub6'],
        'Marks_scored':[89,80,79,97,88,67],
        'roll no':[1,23,4,5,5,6]
    },
    index=[1,2,3,4,5,6])
```

```python
df = pd.merge(three, two, on = 'id', how='right')
df
```

```python
df = pd.merge(three, two, on = 'id', how='left')
df
```

```python
df = pd.merge(three, two, on = 'id', how='inner')
df
```

```python
df = pd.merge(three, two, on = 'id', how='outer')
df
```
