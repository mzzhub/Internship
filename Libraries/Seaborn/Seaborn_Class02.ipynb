# Swarm Plot
* A swarm plot is similar to a strip plot, but it spreads out the points to avoid overlap, giving a better visualization of the distribution of the data.
* Seaborn's `swarmplot()` function is used to create swarm plos.
* Swarmplot is like the scatterplot where points donot overlap.

```python
sns.boxplot(x = "day", y = "total_bill", data = df)
sns.swarmplot(x = "day", y = "total_bill", data = df, color = "0.10")
```

```python
df = sns.load_dataset("tips")
```

```python
sns.boxplot(x = "day", y = "total_bill", data = df)
sns.swarmplot(x = "day", y = "total_bill", data = df, hue = "sex")
```

```python
sns.boxplot(x = "day", y = "total_bill", data = df)
sns.swarmplot(x = "day", y = "total_bill", data = df, hue = "sex", dodge = True)
```

```python
sns.boxplot(x = "day", y = "total_bill", color = "m", data = df)
sns.swarmplot(x = "day", y = "total_bill", data = df, hue = "sex", dodge = True)
```

# Violin Plot
* A violin plot combines the features of a box plot and a kernel density estimation plot to visualize the distribution of a numerical variable across different categories.
* Seaborn's `violinplot()` function is useful for comparing the distribution of numerical variables across categories.

```python
sns.violinplot(data = df, x = "day", y = "total_bill")
```

#Pair Plot
* A pair plot creates a grid of scatter plots for all pairs of variables in a
dataset, along with histograms for each variable's distribution.
* Seaborn's `pairplot()` function can be used to create pair plots.

```python
sns.get_dataset_names()
```
