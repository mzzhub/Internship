# 1. Count all letters, digits, space and special symbols from a string.

```python
string = "A1b@C2d#E3f$G4h%I5j^K6l&M7n*O8p(P9q)R0s+T!u-Vw=Xy_Z"
count = 0
for i in string:
    count += 1
print(count)
# print(len(string))
```

# 2. Replace each special symbol with "\#" in the given string.
`string = "python !$ @ programming !anguage"`

```python
string = "python !$ @ programming !anguage"
output = str()
for i in string:
    if i.isalnum() or i.isspace():
        output += i
    else:
        output += "#"
print(output)
```

# 3. Use a loop to display elements from a given list present at an odd index position.

```python
list01 = list(range(1,16))
print(list01)

index = True
for i in list01:
    if index:
        print(i)
        index = False
    else:
        index = True
```
