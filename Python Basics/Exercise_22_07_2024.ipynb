```python
def validate_pin(pin):
    alpha = 'qwertyuiopasdfghjklzxcvbnmQWERTYUIOPASDFGHJKLZXCVBNM'
    if pin == '':
        print("Invalid, Empty pin is not allowed.")
        return False
    for i in pin:
        if i == " ":
            print("Invalid, Spaces are not allowed.")
            return False
        if i in alpha:
            print("Invalid, Only Numeric Characters are allowed.")
            return False
    pin = int(pin)
    if pin<0:
        print("Invalid, Negative numbers are not allowed.")
        return False
    elif not( (pin>=1000 and pin<=9999) or (pin>=100000 and pin<=999999) ):
        print("Invalid, Pin should consists of exactly 4 or 6 characters.")
        return False
    elif (pin>=1000 and pin<=9999):
        d1 = pin//1000
        d2 = (pin - ( d1*1000 )) // 100
        d3 = (pin - (d1*1000) - (d2*100)) // 10
        d4 = (pin - (d1*1000) - (d2*100) - (d3*10))
        digits = [d1,d2,d3,d4]
        for i in digits:
            if i%2 == 0:
                print("Invalid, Even digits are not allowed.")
                return False
        print("It is a valid Password.")
        return True
    elif (pin>=100000 and pin<=999999):
        d1 = pin//100000
        d2 = (pin - ( d1*100000 )) // 10000
        d3 = (pin - (d1*100000) - (d2*10000)) // 1000
        d4 = (pin - (d1*100000) - (d2*10000) - (d3*1000)) //100
        d5 = (pin - (d1*100000) - (d2*10000) - (d3*1000) - (d4*100)) // 10
        d6 = (pin - (d1*100000) - (d2*10000) - (d3*1000) - (d4*100) - (d5*10))
        digits = [d1,d2,d3,d4,d5,d6]
        for i in digits:
            if i%2 == 0:
                print("Invalid, Even digits are not allowed.")
                return False
        print("It is a valid Password.")
        return True
```
