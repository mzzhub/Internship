```python
def phonebook(pb):
    while True:
        while True:
            operation = (input("1. Add\n2. Update\n3. Delete\n4. View\n5. Quit\nEnter the requried operation: "))
            if operation not in ['1','2','3','4','5']:
                print("Invalid input.\n")
                continue
            operation = int(operation)
            if operation in [1,2,3,4,5]:
                break
        if operation == 1:
            pb = add(pb)

        if operation == 2:
            pb = update(pb)

        if operation == 3:
            pb = delete(pb)

        if operation == 4:
            view(pb)

        if operation == 5:
            break
    if operation == 5:
        print("\nAll changes are saved.")
        return pb



def add(pb):
    print("\nAdding a new Contact.")
    name = input("Enter name: ")
    number = int(input("Enter phone number: "))
    mail = input("Enter email: ")
    print("\nNew contact added successfully.")
    line()
    pb[name] = {"Name":name,"Number":number,"e-mail":mail}
    return pb



def update(pb):
    print("\nUpdating an existing Contact.")
    allcontacts(pb)
    while True:
        row = input("\nEnter the name of contact you want to update: ")
        if row in pb:
            break
        else:
            print("Cannot find the contact.\nTry again")
            continue
    print(pb[row])
    if row in pb:
        new_name = input("Enter new name: ")
        new_number = int(input("Enter new number: "))
        new_mail = input("Enter new e-mail: ")
        print("\nContact updated successfully.")
        line()
        del pb[row]
        pb[new_name] = {"Name":new_name,"Number":new_number,"e-mail":new_mail}
    return pb



def delete(pb):
    print("\nDeleting an existing Contact.")
    allcontacts(pb)
    while True:
        row = input("\nEnter the name of the contact you want to delete: ")
        if row in pb:
            break
        else:
            print("Cannot find the contact.\nTry again")
            continue
    del pb[row]
    print("\nContact has been deleted.")
    line()
    return pb



def allcontacts(pb):
    print("\nAll Contacts.\n")
    for i in pb:
        #print(i,pb[i])
        for j in pb[i]:
            print(j,":",pb[i][j])
        print("\n")



def view(pb):
    allcontacts(pb)
    line()



def line():
    for i in range(149):
        print('-',end="")
    print("\n")
```
