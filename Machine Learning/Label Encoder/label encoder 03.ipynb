```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
outlook_fit = le1.fit(df["Outlook"])
df["Outlook"] = outlook_fit.transform(df["Outlook"])
df
```

```python
le2 = LabelEncoder()
temprature_fit = le2.fit(df["Temprature"])
df["Temprature"] = temprature_fit.transform(df["Temprature"])
df
```

```python
le3 = LabelEncoder()
humidity_fit = le3.fit(df["Humidity"])
df["Humidity"] = humidity_fit.transform(df["Humidity"])
df
```

```python
le4 = LabelEncoder()
windy_fit = le4.fit(df["Windy"])
df["Windy"] = windy_fit.transform(df["Windy"])
df
```

```python
le5 = LabelEncoder()
play_fit = le5.fit(df["Play"])
df["Play"] = play_fit.transform(df["Play"])
df
```
