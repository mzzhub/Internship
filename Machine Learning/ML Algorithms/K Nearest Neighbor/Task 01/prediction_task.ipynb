# **Data Slicing**

## Inputs and Outputs

```python
# Slicing features and labels.

x = df.iloc[:,:8]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
df["diabetes"].value_counts()
```

## Train and Test

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train, x_test, y_train, y_test = tts(x, y, test_size = 0.20, random_state = 0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
y_train.value_counts()
```

# **Standardizing inputs**

```python
from sklearn.preprocessing import StandardScaler as ss
```

```python
scaler = ss()
```

```python
scaler.fit(x_train)
x_train = scaler.transform(x_train)

x_test = scaler.fit_transform(x_test)
```

```python
print(x_train.shape)
print(y_train.shape)
```

# **Model Building**

```python
from sklearn.neighbors import KNeighborsClassifier as knn
```

```python
model = knn(n_neighbors = 5)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
temp = pd.DataFrame(y_pred)
```

```python
temp.value_counts()
```

# **Model Evaluation**

```python
from sklearn.metrics import accuracy_score as accs
```

```python
score = accs(y_test, y_pred)
score
```

```python
from sklearn.metrics import confusion_matrix as cm
```

```python
c_m = cm(y_test, y_pred)
print(c_m)
```

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
m_s_e = mse(y_test, y_pred)
print(m_s_e)
```

```python
from sklearn.metrics import precision_score as ps
```

```python
p_s01 = ps(y_test, y_pred, average = "binary")
p_s01
```

```python
p_s02 = ps(y_test, y_pred, average = "micro")
p_s02
```

```python
from sklearn.metrics import recall_score as rs
```

```python
r_s = rs(y_test, y_pred, average = "binary")
r_s
```
