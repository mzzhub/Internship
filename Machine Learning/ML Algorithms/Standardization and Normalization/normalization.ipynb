* Normalization is also catted MinMaxscalar.
* Class used for normalisation called MinMaxscalar.

```python
print(train_input)
train_input = train_input.reshape(-1, 1)
print(train_input)
print()
scaler.fit(train_input)
print(train_input)
print()
scaler.transform(train_input)
print(train_input)
```
