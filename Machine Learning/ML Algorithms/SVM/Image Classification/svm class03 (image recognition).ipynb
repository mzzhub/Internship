```python
from skimage.io import imread
from skimage.transform import resize
```

```python
flatten_data01 = list()
target01 = list()
path_of_folders = [cat_path, dog_path]

count01 = 0
for folder in path_of_folders:
    print("begin")
    for image in os.listdir(folder):
        image_path = os.path.join(folder, image)
        image3D = imread(image_path)
        resized_image = resize(image3D, (150, 150, 3))
        flattened_image = resized_image.flatten()
        flatten_data01.append(flattened_image)
        target01.append(path_of_folders.index(folder))
    print("end")
```

```python
len(flatten_data01)
```

```python
len(target01)
```

```python
import numpy as np
import pandas as pd
```

```python
flatten_data = np.array(flatten_data01)
target = np.array(target01)

df = pd.DataFrame(flatten_data)
df["target"] = target
```

```python
df
```

```python
df.info()
```

```python
x = df.drop("target", axis = 1)
y = df["target"]
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(x, y, test_size = 0.2, random_state = 2)
```

```python
from sklearn.svm import SVC
classifier = SVC()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score as acs
acs(y_test, y_pred)
```

**Metrics Definition**
1. **Precision** - Precision is defined as the ratio of true and false positives.
2. **Recall** - Recall is defined as the ratio of true positives and false negatives.
3. **F1 Score** - The F1 is the weighted harmonic mean of precision score and recall score. The closer the value of the F1 score is to 1.0, the better the expected performance of the model.
4. **Support** - Support is the number of actual occurences of the class in the dataset. It doesnt vary between models, it just
diagnoses the performance evaluation process.
