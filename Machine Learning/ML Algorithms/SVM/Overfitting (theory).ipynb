# **What is Overfitting?**
* When a model performs very well for training data but has poor performance with test data (new data), it is known as overfitting.
* In this case, the machine learning model learns the details and noise in training data, such that it negatively affects the performance of the model on test data.
* Overfitting can happen due to:
    * Low bias
    * High variance
    * Not clean
    * Contains noise (garbage values)
    * The model has a variance
    * The size of the training dataset used is not enough
    * The model is too complex

# **What is Underfitting?**
* When a model that not learned the pattens in the training data well, is unaUe to generalize well on the new data, it is known as underfitting.
* An underfit model has poor performance on the training data
* It will result in unreliable predictions.
* Underfitting occurs due to:
    * High bias
    * Low variance
    * Not cleared
    * Contains noise (garbage values)
    * The model has a high bias
    * The size of the training dataset used is not enough
    * The model is too simple
    * Inappropriate Features: If important feaures are not inclused in model or irrelevant features are included, the model may struggle to capture the true relationship in the data. leading to underfitting.

# **What is a Good Fit in Machine Learnhg?**
* To find good fit model, you need to loook at the performatre of a machine learning model over time with the training data.
* As the algorithm learns over time, the error for the model in the training data reduces, as well as the error in the test
dataset.
* If you train the model for too long, the model may learn the unneessary details and the noise in the training set increases and hence lead to overfitting.
* In order to achieve a good fit, you need to stop the training process at a point the error starts to increase.

# **Bias**
* Bias is a prediction error that is introduced in model to oversimplyfing machine learning algorithms.
* It is the difference between the predicted values and the actual values.

# **Variance**
* If the machine learning model performs well with the training dataset, but does not perform well with the test dataset, then variance occures.
