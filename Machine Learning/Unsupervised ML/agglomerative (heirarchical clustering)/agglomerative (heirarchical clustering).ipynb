```python
from sklearn.preprocessing import normalize
temp = normalize(df)
```

```python
df_normal01 = pd.DataFrame(temp, columns = df.columns)
```

```python
df_normal01
```

```python
df_normal01.info()
```

```python
import scipy.cluster.hierarchy as sch
dend = sch.dendrogram(sch.linkage(df_normal01, method = "average"))
```

```python
from sklearn.cluster import AgglomerativeClustering as ac
cluster = ac(n_clusters = 2, metric = "euclidean", linkage = "complete")
y = cluster.fit_predict(df_normal01)
y
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 10))
plt.scatter(df_normal01["Milk"], df_normal01["Grocery"], c = y)
plt.xlabel("Milk")
plt.ylabel("Grocery")
plt.show()
```
