```python
from sklearn.cluster import KMeans
```

```python
km01 = KMeans(n_clusters = 3, init = "k-means++", random_state = 0)
km01.fit(df)
```

```python
km01.inertia_
```

```python
# Looping

inertia_list = list()

for i in range(1, len(df) + 1):
    km_temp = KMeans(n_clusters = i, random_state = 0)
    km_temp.fit(df)
    inertia_list.append(km_temp.inertia_)

inertia_list
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, len(df) + 1), inertia_list)
plt.xlabel("n_clusters")
plt.ylabel("KMeans.inertia_")
plt.xticks(range(1, len(df) + 1))
plt.show()
```

```python
km02 = KMeans(n_clusters = 2, random_state = 0)
y_pred = km02.fit_predict(df)
y_pred
```

```python
df["cluster"] = y_pred
```
