# **Natural language processing**
* Natural language processing (NLP) is a branch of artificial intelligence (AI) that enables computers to comprehend, generate, and manipulate human language.
* Natural language processing has the ability to interrogate the data with natural language text or voice.
* NLP has been around for more than 50 years and has many real-world applications, including in business intelligence, search engines, and medical research.

```python
import nltk
```

```python
nltk.download("stopwords")
```

```python
# from nltk.corpus import stopwords
# temp = stopwords.words("english")

stopwords = nltk.corpus.stopwords.words("english")
print(stopwords)
```

```python
nltk.download("punkt_tab")
```

```python
sent01 = '"at eight o clock on thursday morning, hello how are you. I love learning Python programming. I am a programmer'
tokens = nltk.word_tokenize(sent01)
tokens
```

```python
words = [w.lower() for w in tokens if w not in stopwords]
words
```

# **N-grams**
* They are a contiguous sequence of $n$ items (words, characters, or tokens) extracted from a given text or dataset.
* They are widely used in natural language processing (NLP) and text analysis to represent patterns and relationships in data.
* Example :

```python
from nltk.util import ngrams
trigram = list(ngrams(tokens, 3))
trigram
```

# **Stemming**

```python
# Checking SnowballStemmer

from nltk.stem import SnowballStemmer
sbs = SnowballStemmer("english")

for w in words:
    print(w, "\u2192", sbs.stem(w))
```

```python
# Checking PotterStemmer

from nltk.stem import PorterStemmer
ps = PorterStemmer()

for w in words:
    print(w, "\u2192", ps.stem(w))
```

# **Lemmatization**
Lemmatization is the process of reducing a word to its base or root form (lemma) by considering its context and part of speech (POS). It is more sophisticated than stemming as it uses a lexical database like WordNet.

## **Steps to Perform Lemmatization** in Python
1. Import Required Libraries:

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import nltk
nltk.download('wordnet')  # Required for WordNet
```

2. Initialize the Lemmatizer:

```python
lemmatizer = WordNetLemmatizer()
```

3. Perform Lemmatization:



```python
# Without specifying POS (default is noun 'n'):
print(lemmatizer.lemmatize("running"))  # Output: running
```

```python
# By specifying POS:
print(lemmatizer.lemmatize("running", pos="v"))  # Output: run
```

image.png

## **Examples**

```python
# Lemmatizing individual words
print(lemmatizer.lemmatize("geese"))           # Output: goose
print(lemmatizer.lemmatize("better", pos="a")) # Output: good
print(lemmatizer.lemmatize("running", pos="v"))# Output: run
```

```python
# Lemmatizing a list of words
words = ["cats", "geese", "running", "easily", "fairly"]
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
print(lemmatized_words)  # Output: ['cat', 'goose', 'running', 'easily', 'fairly']
```
