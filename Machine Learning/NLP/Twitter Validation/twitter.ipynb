```python
# Droping rows which are "Irrelevant" in "target" column

idx = df.index[(df["target"] == "Irrelevant")]
idx
```

```python
# maping values in column "target"

# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()
# df["target"] = le.fit_transform(df["target"])

df["target"] = df["target"].map({"Negative" : -1, "Neutral" : 0, "Positive" : 1})
```

```python
df["target"].value_counts()
```

```python
tweets = df["text"]
tweets
```

```python
tweets = tweets.str.replace(r"[^a-zA-Z0-9]", " ", regex = True)
```

```python
tweets
```

```python
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer

ss = SnowballStemmer("english")
tt = TweetTokenizer()

tweets = tweets.apply(lambda x : [ss.stem(i.lower()) for i in tt.tokenize(x)]).apply(lambda y : " ".join(y))
```

```python
tweets
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download("stopwords")
```

```python
sw = set(stopwords.words("english"))
```

```python
tweets = tweets.apply(lambda x : [i for i in tt.tokenize(x) if i not in sw]).apply(lambda y : " ".join(y))
```

```python
tweets
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()

train_vec = tfidf.fit_transform(tweets)
```

```python
print(train_vec[0])
```

```python
train_vec.shape
```

```python
x = train_vec
y = df["target"].values
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(x, y, test_size = 0.3, random_state = 0)
```

# **Model Building**

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

svm = SVC()
m_nb = MultinomialNB()
rfc = RandomForestClassifier()
abc = AdaBoostClassifier()

model_list = [svm, m_nb, rfc, abc]
```

```python
from sklearn.metrics import accuracy_score as acs, confusion_matrix as cm, classification_report as cr
```

```python
output01 = list()
for model in model_list:
    temp = []
    temp.append(str(model))
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    temp.append(acs(y_test, y_pred))
    temp.append(cm(y_test, y_pred))
    temp.append(cr(y_test, y_pred))
    output01.append(temp)
```
