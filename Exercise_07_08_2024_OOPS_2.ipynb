# Answer

```python
class College:
    def __init__(self):
        self.college_name = input("Enter College name: ")
        self.dep = input("Enter Department of College: ")
    def mycollege(self):
        print("College:",self.college_name + "\nDepartment:",self.dep)
```

```python
class School:
    def __init__(self):
        self.school_name = input("Enter School name: ")
        self.place = input("Enter location on School: ")
    def my_school(self):
        print("School:",self.school_name + "\nLocation:",self.place)
```

```python
class Student(College, School):
    def __init__(self):
        self.student_name = input("Enter name of the Student: ")
        self.age = int(input("Enter age of the Student: "))
        School.__init__(self)
        College.__init__(self)
    def new(self):
        print("Student Name:",self.student_name + "\nAge:",str(self.age) + "\nSchool:",self.school_name + "\nLocation:",self.place + "\nCollege:",self.college_name + "\nDepartment:",self.dep)
```
